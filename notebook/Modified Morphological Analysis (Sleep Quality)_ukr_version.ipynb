```python
# Function to handle missing pairs by returning a default value (e.g., 0)
def get_p_ij(val1, val2):
    pij_value = pij.get((val1, val2))
    if pij_value is None:
        #print(f"Warning: get_p_ij returned None for {val1} and {val2}")
        return 0  # Default value for undefined pairs
    return pij_value

# # Create value pairs for interactions between column values
# def create_value_pairs(row, columns):
#     pairs = list(itertools.combinations(columns, 2))
#     return [f"{row[col1]} {row[col2]}" for col1, col2 in pairs]

# Calculate the cumulative product of interactions for each row
pairs = list(itertools.combinations(columns, 2))

def calculate_c(row):
    c_value = 1
    for col1, col2 in pairs:
        p_ij = get_p_ij(row[col1], row[col2])
        c_value *= (1 + p_ij)
    return c_value

df['Сукупність C'] = df.apply(calculate_c, axis=1)

# Show the final output
df.iloc[2000:2030,:]
```

```python
df['P*C'] = df['Сукупність P']*df['Сукупність C']
```

```python
#Розраховані на першому етапі МММА оцінки альтернатив

df['P*C normalised'] = df['P*C'] / df['P*C'].sum()
df.iloc[1300:1320,:]
```
